### E-Commerce Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
csv=pd.read_excel(r"C:\Users\91638\Documents\Superstore.xlsx")
print(csv)
```

```python
csv.describe()
 
```

```python
### sort values
csv.sort_values(["Profit","Sales"],axis=0,ascending=[False,False],inplace=True)
```

```python
# Graph b/w region and category
fig= px.bar(csv.head(20),x='Region',color='Category',width=700 , title='Graph b/w region and category')
fig.show()
```

```python
#graph b/w region and profit
fig= px.bar(csv.head(20),x='Region',y="Sales",color='Profit',width=700,title='Sales in Region with heighest value')
fig.show()
```

```python
##grapg b/w order count vs sub category
fig=px.bar(csv.head(100),x='Sub-Category',color='Category',width=700,title='order count of category sub-category')
fig.show()
```

```python
fig=px.bar(csv.head(20),x='Customer Name',y='Quantity',color ='Profit',hover_data=['Discount','Category'],width=700, title='order count of category sub-category')
fig.show()
```

```python
fig=px.pie(csv,names='Sub-Category',values='Profit',width=700,title='order count of category sub-category')
fig.show()
```
